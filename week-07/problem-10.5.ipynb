```python
import numpy as np
import matplotlib.pyplot as plt 
from numpy.random import default_rng
rand = default_rng().random
```

```python
#testing the Gillespie Algorithm
l0 = 0 
beta_birth = 1 
k_death = 0.3

N = 100

l = l0
T = 0
l_r = np.zeros(N)
t_r = np.zeros(N)
for n in range(N):
    rate = beta_birth+ k_death*l
    t = - 1/rate * np.log(rand())
    xi = beta_birth/rate
    event = rand() < xi
    if event: 
        l = l+1
    else: 
        l = l-1
    T = T+t
    l_r[n] = l
    t_r[n] = T
    plt.plot(T,l,'.')
```

```python
#Developing a function for the Gillespie Algorithm
def gillespie(l_init,T):
    t_s = np.zeros(T)
    l_s = np.zeros(T)
    beta_birth = 1
    k_death = 0.3
    N = T
    T0 = 0 
    l = l_init
    
    for n in range(N):
        rate = beta_birth + k_death*l
        t = - 1/rate * np.log(rand())
        xi = beta_birth/rate
        event = rand() < xi
        if event: 
            l = l+1
        else: 
            l = l-1
        T0 = T0+t
        t_s[n] = T0
        l_s[n] = l 
    
    return t_s, l_s
```

```python
num_sim = 150
T = 300
l0 = 0 
l = np.empty((150,300))
time = np.empty((150,300))
for i in range(num_sim):
    res = gillespie(l0,T)
    l[i] = res[1]
    time[i] = res[0]

```

```python
sum_l = np.ones(T)
sum_t = np.ones(T)
for n in range(T):
    sum_l[n] = l[0:,n].sum() 
    sum_t[n] = time[0:,n].sum()
average_l = sum_l/num_sim
average_t = sum_t/num_sim
```

```python
plt.plot(average_t,average_l)
```
