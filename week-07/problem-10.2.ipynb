```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
data = np.load('colquhounData.npy')
split = np.hsplit(data,2)
bins = split[0].flatten()
channels = split[1].flatten()
```

## (a)

```python
N = channels.sum()
open_channels = channels.cumsum()
print(N)
print(open_channels)
```

## (b)

```python
plt.figure()
plt.plot(bins,channels,label='closed channels')
plt.title('Number of Closed Channels')
plt.figure()
plt.plot(bins,open_channels,label='open channels')
plt.title('Number of Open Channels')
```

The plot of closing channels looks slightly nosier than the open channels. This is possibly due to the randomness in which the channels reopened after being exposed to the neurotransmitter.
