## (a) Dimensional analysis

- What combination of $m$, $g$, and $\zeta$ defines the timescale $T$?

we can define $T$ as $T = \dfrac{m}{\zeta}$ giving the dimension of $\dfrac{1}{s}$ We can then define a new parameter $\overline{t} = \dfrac{t}{T}$

- What does the differential equation look like when we rescale $t$ by $T$?

Rescaling $t$ by $T$ yields a nondimensional equation of the form $\dfrac{d^2\theta}{d\overline{t}^2} = -(\dfrac{gT^2}{L}) sin(\theta) - \dfrac{d\theta}{d\overline{t}}$

- What does the differential equation look like when we make the small angle approximation $\sin \theta \approx \theta$?

Utilizing a small angle approximation the differential equation becomes: $\dfrac{d^2\theta}{d\overline{t}^2} = -(\dfrac{gT^2}{L})\theta - \dfrac{d\theta}{d\overline{t}}$ which can then be further reduced to only include a single adjustable paramenter $\alpha$ resulting in the following: $\dfrac{d^2\theta}{d\overline{t}^2} = -\alpha \theta - \dfrac{d\theta}{d\overline{t}}$

- What is the solution of this approximate differential equation?

The approximate solution of this equation would result in a sinusoid of some sort. 

## (b) Classify solutions

What criterion determines whether solutions will overshoot the stable point or not?

```python
def F(y,t):
    dy = [0,0]
    dy[0] = y[1]
    dy[1] = -(alpha*y[0] + y[1])
    return dy
```

```python
m = 2
g = 9.81
L = 2
zeta = 1
T = zeta/m
alpha =(g*T**2)/L
t = np.linspace(0,10,31)
y0 = [1,4]
y = odeint(F,y0,t)
plt.plot(t,y[:,0])
plt.plot(t,y[:,1])
```

Adjusting zeta changes the system response and changes the type of damping that occurs within the system.
