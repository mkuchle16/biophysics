```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

## (a) Plot the phase portrait.

```python
def F(y,t):
    dy = [0,0]
    dy[0] = y[1]
    dy[1] = -(m*g*np.sin(y[0]) + delta*L*y[1])/(m*L)
    return dy
```

```python
m = 2
g = 9.81
L = 1
delta = 1
t = np.linspace(-4*np.pi,4*np.pi,201)
y0 = [1,4]
y = odeint(F,y0,t)
test1 = y
```

```python
y0 = [-1,3]
y = odeint(F,y0,t)
test2 = y
```

```python
y0 = [1,5]
y = odeint(F,y0,t)
test3 = y
```

```python
y0 = [-3,0]
y = odeint(F,y0,t)
test4 = y
```

```python
X,Y = np.meshgrid(y[:,0],y[:,1])
vX = Y
vY = -X
plt.figure(figsize=(16,9))
plt.quiver(X,Y,vX,vY,pivot = 'mid')
plt.plot(test1[:,0],test1[:,1])
plt.plot(test2[:,0],test2[:,1])
plt.plot(test3[:,0],test3[:,1])
plt.plot(test4[:,0],test4[:,1])
plt.axhline(y=0,color='red')
plt.plot(0,0,'or',ms=5)
```
