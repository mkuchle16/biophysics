```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

## (a) Plot the phase portrait.

```python
def F(y,t):
    dy = [0,0]
    dy[0] = y[1]
    dy[1] = -(m*g*np.sin(y[0]) + delta*L*y[1])/(m*L)
    return dy
```

```python
m = 2
g = 9.81
L = 1
delta = 1
t = np.linspace(-2*np.pi,2*np.pi,101)
y0 = [1,4]
y = odeint(F,y0,t)
```

```python
plt.plot(t,y[:,0])
plt.plot(t,y[:,1])
```

```python
X,Y = np.meshgrid(y[:,0],y[:,1])
vX = Y
vY = -X
plt.figure(dpi = 200)
plt.quiver(X,Y,vX,vY,pivot = 'mid')
plt.plot(y[:,0],y[:,1])
plt.axhline(y=0,color='red')
plt.plot(0,0,'or',ms=5)
```

```python
# Just a test for plt.quiver
x = np.linspace(-np.pi,np.pi,21)
y = np.linspace(-np.pi,np.pi,21)
X,Y = np.meshgrid(x,y)
uX = Y
uY = -X
plt.quiver(X,Y,uX,uY,pivot='mid')
plt.plot()
```
