```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

## (a) Plot the phase portrait.

```python
def F(y,t):
    dy = [0,0]
    dy[0] = y[1]
    dy[1] = -(m*g*np.sin(y[0]) + delta*L*y[1])/(m*L)
    return dy
```

```python
m = 2
g = 9.81
L = 1
delta = 0
t = np.linspace(-1,1,31)
y0 = [10,4]
y = odeint(F,y0,t)
```

```python
plt.plot(t,y[:,0])
plt.plot(t,y[:,1])
```

```python
X,Y = np.meshgrid(y[:,0],y[:,1])
vX = Y
vY = -X
plt.figure(figsize=(16,9))
plt.quiver(vX,vY,pivot = 'mid')
```
