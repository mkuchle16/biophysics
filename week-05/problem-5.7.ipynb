```python
#Setup of necessary packages
import numpy as np 
import matplotlib.pyplot as plt

from numpy.random import default_rng
rng = default_rng()
rand = rng.random
```

## (a) __Histogram of $1/x$__

```python
x = np.random.uniform(0,1,10000)
inv_x = 1/x
count, bins, ignored = plt.hist(inv_x,30, density = True, range = (0,10))
```

## (b) Log-log plot of frequencies

```python
plt.loglog(count)
```

## (c) Analysis and Conclusions

The loglog plot of this distribution indicates a power law relationship between the observered frequency. Initially the histogram would possibly indicate exponential decay in the numbers, or some exponential relationship like a poisson distribution. This distribution makes sense becasue we took the reciprocal of the values creating a new power law relationship from the uniform distribution. 

## (d) Numerical experiments with more samples

```python
#Replacing 10000 tests with 50000 tests an anlyzing results
x = np.random.uniform(0,1,50000)
inv_x = 1/x
count, bins, ignored = plt.hist(inv_x,100, density = True, range = (0,10))
```

```python
plt.loglog(count)
```

The test with `10000` samples shows some slight noise at the end of the loglog plot as x seems to get significantly large. With `50000` samples this effect seems to become amplified and the noise or randomness seems to increase significantly possibly indicating that the power law relationship becomes less of an effective model as the value of x becomes very large or the values of inv_x become increasingly more random.
